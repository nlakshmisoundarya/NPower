````python
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
````

````python
# Import the sales data
df = pd.read_csv('company_sales_data.csv')
df.head()
````

````python
# Check for size of data, data type, and if any data is missing
df.info()
````

## 1. Create a basic line plot of tot_profit as a function of month_number.

````python
# Uncomment the following code to make a basic plot using pyplot from matplotlib:

plt.plot(df['month_number'], df['total_profit'])
plt.show() # Need this line to render the plot!
````

Notice how bare that plot is? There are lots of ways to adjust plots using matplotlib to change plots to improve clarity, highlight key insights, and improve the aesthetics of the visual.
# 2. Add and alter some of the key components that make up the anatomy of a plot.

Uncomment the key components one at a time while re-executing this cell to see how the plot changes! \
Be sure to change the inputs to improve the graph and better describe the data.

````python
# Create the same basic plot to start
plt.plot(df['month_number'], df['total_profit'])

# # Add a title to the plot:
plt.title('Descriptive title here')

# # Add x and y axis labels:
plt.xlabel('month_number')
plt.ylabel('total_profit')

# # Starting from 0 is often important so as to not mislead the audience.
# # Change the limits of the axes to frame the data:
plt.xlim(1,12)
plt.ylim(0,500000)

# # Change the tick marks of the x and y axis:
plt.xticks(range(1,13))
plt.yticks(range(0, 550000, 50000))

# # Add grid lines to better read values off the graph.
plt.grid()

# # Add a legend:
# plt.legend(['Trend Line Label'], loc='lower right', fontsize=12)
# # Try to move the legend to the upper left!
plt.legend(['Trend Line Label'], loc='upper left', fontsize=12)

# Render the plot
plt.show()
````

# 3. Improve the utility of the figure.
We will now add mutliple trend lines and plot points by applying more than one plot method in the same figure, as well as customizing line plots, tweaking the axis labels with custom strings, and adding annotations.

````python
# Create the same basic plot to start

# Change the line style (ls), line width (lw) and color (c):
# Use the label argument to set what will display in the legend:
plt.plot(df['month_number'], df['total_profit'], c='g', ls='--', lw=2, label = 'Total Profit', marker='s')

# Add a trend line to highlight the mean profit:
# Move it behind the profit trend line by placing this line first in the cell:
plt.plot(df['month_number'], np.ones(12)*df['total_profit'].mean(), c='b', ls='--', lw=1, label='Avg. Profit')

# Add markers to highlight the maximum and minimum:
# Set the marker style (marker), marker size (ms), and color(c):
plt.plot(df.loc[df['total_profit'].idxmax(), 'month_number'], df['total_profit'].max(), c='r', marker='o', ms=8)
plt.plot(df.loc[df['total_profit'].idxmin(), 'month_number'], df['total_profit'].min(), c='r', marker='o', ms=8)

# Control the size of titles and labels with the fontsize argument:
# Add a title to the plot:
plt.title('Yearly Profit', fontsize=14)

# # Add x and y axis labels:
plt.xlabel('Month', fontsize = 12)
plt.ylabel('Profit [k$]', fontsize = 12)

# Starting from 0 is often important so as to not mislead the audience.
# Change the limits of the axes to frame the data:
plt.xlim(1,12)
plt.ylim(0,500000)

# Change the tick marks of the x and y axis:
# Now add more useful label names by providing lists:
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
profit_in_thousands = list(range(0, 550, 50))
plt.xticks(range(1,13), months)
plt.yticks(range(0, 550000, 50000), profit_in_thousands)

# Add grid lines to better read values off the graph.
plt.grid()

# Add a legend:
plt.legend(loc='upper left', fontsize=12)

# Add annotations:
plt.annotate(
    text='So Much Profit',  # Annotation text
    xy=(11, df['total_profit'].max()),   # Point to annotate (data coordinates)
    xytext=(7, 405000),  # Position of the text (data coordinates)
    arrowprops=dict(facecolor='black', shrink=0.05), # Arrow properties
    fontsize=12
)

# Add an extra annotations:
plt.annotate(
    text='Extra text',  # Annotation text
    xy=(3, 60000),   # Point to annotate (data coordinates)
    fontsize=12
)


# Render the plot
plt.show()
````

# 4. Your turn! Make a line plot of total_units as a function of month_number. How does it compare to profit over that same time?

````python
# Make your plot here:
plt.plot(df['month_number'], df['total_units'], color='b', linestyle='-', linewidth=2, label = 'Total Units Sold', marker='o')
plt.title('Monthly Total Units Sold', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.grid()
plt.xticks(range(1,13), months)
# Make sure to render the plot!
plt.show()
````

# 5. Create a bar chart for total yearly item sales to see which product generated the largest revenue:

````python
# Make your bar chart here:
yearly_sales = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
plt.barh(yearly_sales.index, yearly_sales.values, color = 'r')
plt.title("Total yearly sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()

# Make sure to render the plot!
# plt.show()
````
