# DataFrame Attributes

```python
d=[{'Name':'Geetu','Roll':1},{'Name':'Aaliyah','Roll':2},{'Name':'Ananya','Roll':3}]
df = pd.DataFrame(d,index=['A','B','C']) #A,B,C as index
# df = pd.DataFrame(d)  --->0,1,2 as indes
df
```

```python
df.size  #total number of entries  rows*columns
```

```python
df.shape #dimesnion (rows, columns)
```

```python
df.axes #resturns a list representing axes of dataframe
```

```python
df.ndim  #number represesnts the number of axes/array dimention 
```

```python
df.columns  #column names in a list
```

```python
df.values  #row enteies in a list each
```

```python
df['Grade']=['A','B','C']  #adding new column to a dataframe
df
```

```python
#df.loc[:]
# select all rows (like slicing)   ---   Equivalent to df.loc[:, :] (all rows, all columns)

df.loc[:,"Grade2"]=['A','B','C']
df
```

```python
#df.loc[int]
# select a specific row by index label  --- Returns a single row as a Series
df.loc[2]=["Bhoomi","Social","D","D"]
df
```

```python
# drop() by default returns a new DataFrame without modifying the original df unless you:

# Reassign it, or
# Set inplace=True   ---->   # df.drop('Grade',axis=1,inplace=True)
# axis=1 means operate on columns
# axis=0 means operate on rows

df =df.drop('Grade',axis=1)
df
```

```python
df.drop(2,axis=0,inplace=True)
```
