READING DATA IN PANDAS

```python
#reading headers
df.columns

#reading specific column
df['Name'][:5] #[0] to select which row or [0:5] no. of rows we want
#if we want specific multiple columns
df[['Name','Type 1','HP']]

#reading each row
df.iloc[:1] #iloc means integer location like [1], and give the row number that we want, [1:4] multiple rows

#reading from a specific location (R,C)
df.iloc[2,1]

#we can even loop
for index, row in df.iterrows():
  print(index, row['Name'])

#to find data locations with some specific data
df.loc[df.Name =='Bulbasaur'] #or df.loc[df['Name']=='Bulbasaur']
```

SORTING/DESCRIBING DATA

```python
df.sort_values('Name',ascending=False)
#for multiple
df.sort_values(['Name','Type 1','Defense'], ascending=False)
```

```python
df.head(1)
```

MAKING CHANGES TO DATA

```python
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Speed']+df['Sp. Atk']+df['Sp. Def']
df.head()
```

```python
#Now after adding that till we drop total it stays so to drop a specific column
df.drop(columns ='Total')

#Now to add it again
df['Total'] = df.iloc[:,4:10].sum(axis=1) #: says all rows, 4th to 10th columns,
#we want sum so .sum axis =1 says vertically axis=0 is horizontally

cols = list(df.columns.values)
df = df[cols[:4]+[cols[-1]]+cols[4:12]] #since df[-1] is single column we should make it a lst first to add so [df[-1]]

df.head()
```

TO SAVE OUR NEW FILE AFTER MAKING DESIRED CHANGES

```python
df.to_csv('Modified.csv',index=False) #we will get in our files
#if we dont want to have indexes keep index =false

#to have in excel format
df.to_excel('Modified.xlsx',index=False)

#to have in text format
df.to_csv('Modified.txt', index=False, sep='\t')
```

**FILTERING OUR DATA**

Method	        --->   	           When to Use

df.loc[]	    --->        Label-based filtering & selection

df.iloc[]		  --->          Position-based filtering & selection

df.query()		--->          SQL-like filtering

df[df['col'] == value]--->	Quick row filtering

df.at[]		    --->          Fast single value access (by label)

df.iat[]		  --->          Fast single value access (by position)

df.where()		--->          Keeps original shape with NaN

df.mask()		  --->         Replaces matching values with NaN
